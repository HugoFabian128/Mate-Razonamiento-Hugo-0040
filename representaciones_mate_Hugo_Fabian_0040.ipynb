#### Problema 2
Problema 2: Edades de Pedro y AnaPaso 1: Análisis y resolución algebraicaDefinir las variables:Edad de Ana: $x$ (es nuestra referencia más simple).Edad de Pedro: $2x$ (ya que tiene el doble que Ana).Plantear la ecuación: La suma de ambas edades es 36.$$x + 2x = 36$$Simplificar: Sumamos los términos semejantes ($1x + 2x = 3x$).$$3x = 36$$Despejar la variable: El 3 que multiplica a $x$ pasa dividiendo al otro lado.$$x = \frac{36}{3}$$$$x = 12$$Resultado final:Ana ($x$): 12 años.Pedro ($2x$): $2 \times 12 = 24$ años.

```python
# Problema 2: Pedro tiene el doble que Ana y suman 36
# suma_total = 36
suma_total = int(input("Ingrese la suma total: "))
# La ecuación es x + 2x = 36 -> 3x = 36
# Por lo tanto, x (Ana) es suma_total dividido entre 3
edad_ana = suma_total / 3
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
